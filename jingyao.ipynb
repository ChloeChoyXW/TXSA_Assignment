```python
import pandas as pd

# load data
df = pd.read_csv('preprocessed_cyberbullying_tweet.csv')
df.head(10)
```

```python
# Function to convert lists to strings
def convert_list_to_string(lst):
    return ' '.join(lst) if isinstance(lst, list) else lst

# Apply the function to the 'text' column
df['tweet_text'] = df['tweet_text'].apply(convert_list_to_string)
```

```python
import numpy as np
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
# tokenize text
def tokenize_text(text):
    if isinstance(text, str):  # Check if text is a string
        tokens = word_tokenize(text)
        return tokens
    else:
        return []  # Return an empty list for NaN values

# Apply tokenization to the text column in the DataFrame
df['tweet_text'] = df['tweet_text'].apply(tokenize_text)
```

```python
from nltk.corpus import words
import nltk

# Download the list of English words (if not already downloaded)
nltk.download('words')

# Load the set of English words
english_words = set(words.words())

# Function to remove non-English words from a list of tokens
def remove_non_english(tokens):
    english_tokens = [
        token if (token in english_words) else ''
        for token in tokens]
    return [token for token in english_tokens if token != '']

# Apply the function to the 'text' column in the DataFrame
df['tweet_text'] = df['tweet_text'].apply(remove_non_english)
df.head(10)
```

```python

# Remove words equal to or less than 2 characters in length
def remove_short_words(text):
    return ' '.join([word for word in text.split() if len(word) > 2])
 
df['tweet_text'] = df['tweet_text'].apply(remove_short_words)
 
```
